```python
# matplotlib
%matplotlib inline 
#그래프를 브라우저에 직접나타냄
n = 10
y = np.random.randn(n)
plt.plot(range(n),y)
plt.xlabel("시간")
plt.ylabel("랜덤값")
```

```python
#랜덤숫자
np.random.rand(3) # 0~1 숫자 3개

#몬테카를로 분석
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)

plt.figure(figsize=(6,6))
plt.scatter(x,y,s=1)
xx =np.linspace(0,1,100) # 0~1 100개 x값
plt.plot(xx, (1-xx*xx)**0.5,c='r')
```

```python
# 주사위 시행
np.random.randint(1,7,10)
x = np.random.randint(1,7,10000)
plt.hist(x,bins=6, width=0.6)
plt.show()

#정규분포
z = 0
for i in range(2):
    x = np.random.randint(1,7,10000)
    z = z+x
plt.hist(z,bins=11,width=0.6)
plt.show()

# 표준 정규분포 randn
np.random.randn(3) # 평균0 표준편차 1

x = np.random.randn(100000)
plt.hist(x,bins=100,width=0.03)
plt.show()
```
